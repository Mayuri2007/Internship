```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([23, 34, 42, 49, 55, 61, 68, 73, 78, 82])
y = np.array([25, 29, 33, 37, 44, 47, 57, 62, 67, 71])
z = np.array([87, 82, 78, 73, 67, 63, 55, 51, 47, 39])
```

```python
# LINE PLOT
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 5);
plt.grid()

# markers => >,<,^,H,o,*
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 5);
plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 5, label = '2023');
plt.plot(x, z, marker = 'o', mfc = 'yellow', color = 'black', ls = '--', ms = 5, label = '2024');
plt.grid()
plt.legend()
```

```python
# SUB PLOT
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 5);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = 'o', mfc = 'yellow', color = 'black', ls = '--', ms = 5);
plt.grid()
```

```python
# BOX PLOT
```

```python
marks = [67, 65, 55, 59, 69, 72, 51, 70, 53, 98, 23]
```

```python
plt.boxplot(marks);
```

```python
q3=np.percentile(marks, 75)
q1=np.percentile(marks, 25)
q3,q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3 + (1.5 * iqr)
lb = q1 - (1.5 * iqr)
```
