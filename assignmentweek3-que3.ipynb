```python
q3=np.percentile(df['Year'],75)
q1=np.percentile(df['Year'],25)
q3,q1
```

```python
iqr=q3-q1
iqr
```

```python
ub=q3+(1.5*iqr)
lb=q1-(1.5*iqr)
ub,lb
```
